```python
import json
with open("overlapped_frequency_experiment.json", "r+") as f:
    data = json.load(f)

data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 3))  # smaller, wide figure

plt.step(
    data["requested_frequency_change_time_stamps"],
    data["requested_frequency_changes"],
    where="post",
    label="Requested GPU Frequency",
    linewidth=2
)

plt.step(
    data["captured_frequency_time_stamps"],
    data["captured_frequency_changes"],
    where="post",
    label="Observed GPU Frequency",
    linewidth=2,
    linestyle="--"
)

plt.title("Requested vs Observed GPU Frequency Over Time")
plt.xlabel("Time (s)")
plt.ylabel("GPU SM Frequency (MHz)")

plt.legend(loc="best", frameon=False)
plt.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
